### About Dataset

Link dataset: https://www.kaggle.com/datasets/lainguyn123/student-performance-factors


Dataset **"Student Performance Factors"** adalah sekumpulan data yang berisi informasi mengenai berbagai faktor yang berpengaruh terhadap kinerja akademik siswa, dengan fokus pada skor ujian akhir. Berikut adalah penjelasan dari setiap fitur dalam dataset ini:

1. **Hours_Studied**: Jumlah jam yang dihabiskan untuk belajar setiap minggu. Faktor ini mengukur seberapa banyak waktu siswa berinvestasi dalam belajar.
2. **Attendance**: Persentase kehadiran siswa dalam kelas. Faktor ini melihat seberapa sering siswa hadir di kelas, yang mungkin memengaruhi pemahaman materi.
3. **Parental_Involvement**: Tingkat keterlibatan orang tua dalam pendidikan siswa, dengan tiga kategori: Low (rendah), Medium (sedang), dan High (tinggi). Ini menilai peran dukungan orang tua.
4. **Access_to_Resources**: Ketersediaan sumber daya pendidikan, misalnya buku, komputer, atau perpustakaan, dengan kategori Low, Medium, dan High.
5. **Extracurricular_Activities**: Partisipasi dalam kegiatan ekstrakurikuler (Ya atau Tidak). Faktor ini mengukur dampak aktivitas di luar kelas terhadap performa akademik.
6. **Sleep_Hours**: Rata-rata jumlah jam tidur per malam. Tidur memengaruhi tingkat konsentrasi dan kemampuan belajar siswa.
7. **Previous_Scores**: Skor dari ujian sebelumnya. Ini adalah indikator kinerja akademik masa lalu siswa.
8. **Motivation_Level**: Tingkat motivasi siswa dengan tiga kategori (Low, Medium, High). Tingkat motivasi ini berkaitan dengan keinginan siswa untuk belajar dan mencapai hasil akademik.
9. **Internet_Access**: Ketersediaan akses internet (Ya atau Tidak). Akses internet dapat mempengaruhi kemampuan siswa untuk mengakses materi belajar online.
10. **Tutoring_Sessions**: Jumlah sesi bimbingan belajar yang dihadiri setiap bulan. Faktor ini mengukur dukungan ekstra dalam belajar di luar sekolah.
11. **Family_Income**: Tingkat pendapatan keluarga dengan kategori Low, Medium, dan High. Ini mungkin berkaitan dengan akses siswa terhadap sumber daya pendidikan.
12. **Teacher_Quality**: Kualitas guru dengan kategori Low, Medium, dan High. Kualitas pengajaran dapat memengaruhi pemahaman dan prestasi siswa.
13. **School_Type**: Jenis sekolah yang dihadiri, apakah Public (negeri) atau Private (swasta).
14. **Peer_Influence**: Pengaruh teman sebaya terhadap prestasi akademik siswa, dengan kategori Positive, Neutral, atau Negative.
15. **Physical_Activity**: Rata-rata jumlah jam aktivitas fisik per minggu. Aktivitas fisik bisa mempengaruhi kesehatan dan kesejahteraan siswa.
16. **Learning_Disabilities**: Apakah siswa memiliki gangguan belajar (Ya atau Tidak). Ini adalah faktor penting dalam memahami kesulitan siswa dalam proses belajar.
17. **Parental_Education_Level**: Tingkat pendidikan tertinggi yang dicapai oleh orang tua siswa (SMA, Perguruan Tinggi, atau Pasca Sarjana).
18. **Distance_from_Home**: Jarak rumah dari sekolah dengan kategori Near (dekat), Moderate (sedang), atau Far (jauh). Ini bisa memengaruhi waktu siswa untuk belajar atau beristirahat.
19. **Gender**: Jenis kelamin siswa (Laki-laki atau Perempuan).
20. **Exam_Score**: Skor ujian akhir siswa, yang menjadi variabel target atau hasil akhir yang dipengaruhi oleh faktor-faktor di atas.

Dataset ini sangat berguna untuk menganalisis faktor-faktor yang mempengaruhi prestasi akademik siswa, serta dapat digunakan dalam model prediksi untuk memproyeksikan kinerja akademik berdasarkan variabel-variabel tersebut.

# Function Programming

```python
df = pd.DataFrame(student_data)
df
```

```python
#@title Lamda

# Menggunakan fungsi lambda untuk menghitung rata-rata skor ujian berdasarkan Hours_Studied
average_score_by_study_hours = df.groupby('Hours_Studied').agg(
    Average_Exam_Score=('Exam_Score', lambda x: x.mean())
)

print("Rata-rata Skor Ujian Berdasarkan Jam Belajar:")
print(average_score_by_study_hours)


# Menggunakan lambda if else dan menambahkan kolom baru
mean_exam_score = df['Exam_Score'].mean()
df['Above_Average'] = df['Exam_Score'].apply(lambda x: 'Yes' if x > mean_exam_score else 'No')

print("\nData dengan Kolom Above_Average:")
print(df[['Exam_Score', 'Above_Average']])


# Lambda dengan filter data di kolom exam yang lebih dari 75
high_scores = list(filter(lambda x: x > 75, df['Exam_Score']))
print("Skor ujian di atas 75:", high_scores, "\n")

# Lambda dengan map
incremented_scores = list(map(lambda x: x + 1, df['Exam_Score']))
print("Skor ujian setelah ditambahkan 1:", incremented_scores, "\n")

from functools import reduce

# Lambda dengan reduce
total_score = reduce(lambda x, y: x + y, df['Exam_Score'])
print("Total skor ujian:", total_score)
```

```python
#@title Recursion
def total_exam_score(data, index=0, limit=100):
    # Tetapkan batas default untuk panjang data jika tidak disediakan
    if limit is None:
        limit = len(data)

    # Cek jika index melebihi panjang data atau batas rekursi tercapai
    if index >= len(data) or index >= limit:
        return 0

    # Rekursi untuk menjumlahkan skor ujian
    return data['Exam_Score'].iloc[index] + total_exam_score(data, index + 1, limit)

# Menghitung total skor ujian
total_score = total_exam_score(df)
print(f"\nTotal Skor Ujian: {total_score}")
```

```python
#@title First Class
# Fungsi untuk mengkategorikan hasil ujian
def categorize_exam_score(score):
    if score >= 75:
        return 'High'
    elif score >= 65:
        return 'Medium'
    else:
        return 'Low'

# Fungsi untuk menerapkan fungsi kategorisasi pada kolom
def apply_category_to_exam_scores(df, category_function):
    df['Exam_Score_Category'] = df['Exam_Score'].apply(category_function)
    return df

df = apply_category_to_exam_scores(df, categorize_exam_score)
print("DataFrame dengan Kategori Skor Ujian:")
print(df)
```

### Filter Dataframe

* with logic operator

```python
filter_with_operator = data[data.year >= 2015]
filter_with_operator
```

* Multiple Filter

```python
multiple_filter = data[(data.year >= 2015) & (data.transmission == 'Manual') & (data.price > 13998) & (data.mileage >= 3000) & (data.mpg >= 36.2) ]
multiple_filter
```

* Filter with isin

```python
kolom_year = [2015, 2016, 1017]
tahun_mobil = data[data.year.isin(kolom_year)]
tahun_mobil
```

```python
engineSize_mobil = data[data.engineSize.isin([1.8, 2.0, 2.5])]
engineSize_mobil
```

* Filter with str accessor

```python
data[data.model.str.endswith('ry')]
```

```python
data[data.transmission.str.startswith('Ma')]
```

* Filter with nlargest

```python
#Mengambil 3 data belanja pengguna dengan  total pembelian terbanyak
data.nlargest(5, 'price')
```

* Filter with nsmallest

```python
data.nsmallest(5, 'price')
```

### Group dataframe

```python
# # Mengelompokkan data berdasarkan 'year' dan 'transmission', lalu menghitung rata-rata 'price' dan 'mileage'
df.groupby(['year', 'transmission']).agg({'price': 'mean','mileage': 'mean'})
```

```python
# Group by 'year' dan hitung jumlah mobil per tahun
df.groupby('year')['model'].count()
```

```python
# Group by 'year' dan hitung rata-rata harga
df.groupby('year')['price'].mean()
```

```python
#  Group by 'fuelType' dan hitung rata-rata mileage
df.groupby('fuelType')['mileage'].mean()
```

```python
# Group by 'engineSize' dan hitung rata-rata harga
df.groupby('engineSize')['price'].mean()
```

```python
 df.groupby('fuelType').agg({
    'price': 'sum',       # Jumlahkan harga untuk setiap jenis bahan bakar
    'mileage': 'mean',    # Hitung rata-rata mileage untuk setiap jenis bahan bakar
    'mpg': 'max',         # Ambil mpg tertinggi untuk setiap jenis bahan bakar
    'engineSize': 'min'   # Ambil ukuran mesin terkecil untuk setiap jenis bahan bakar
})
```
